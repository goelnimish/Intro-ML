## Linear Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
mse=cross_val_score(lin_reg, X , y , scoring = 'neg_mean_squared_error', cv=10)
print(mse, '\n \n')
mean_mse=np.mean(mse)
print('Mean MSE:', mean_mse)
```

## Ridge Regression 

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
parameters = {'alpha' : [1e-15,1e-10,1e-8,1e-5,1e-3,1e-2,1,5,10,20,25,30,35,40,45,50,55,100,200,250,300,350,433,500,750,900]}
ridge_regressor = GridSearchCV(ridge,parameters,scoring = 'neg_mean_squared_error', cv=10)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

## Lasso Regression 

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
parameters = {'alpha' : [1e-15,1e-10,1e-8,1e-5,1e-3,1e-2,1,5,10,20,25,30,35,40,45,50,55,100,200,250,300,350,433,500,750,900]}
lasso_regressor = GridSearchCV(lasso,parameters,scoring = 'neg_mean_squared_error', cv=10)
lasso_regressor.fit(X,y)

print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_train_lr = r2_score(y_train, y_pred_train)
r2_test_lr = r2_score(y_test, y_pred_test)

print("r2 Train Linear Reg Score:", r2_train_lr)
print("r2 TestLinear Reg Score:", r2_test_lr)
```

### Since the train, test difference is more that 5% we will say that our model is overfitting

```python
ridge=Ridge(alpha=100)
ridge.fit(X_train, y_train)
print(ridge.intercept_)
print(ridge.coef_)
```

```python
y_pred_train = ridge.predict(X_train)
y_pred_test = ridge.predict(X_test)

from sklearn.metrics import r2_score
r2_train_ridge = r2_score(y_train, y_pred_train)
r2_test_ridge = r2_score(y_test, y_pred_test)

print("r2_train_ridge Score:", r2_train_ridge)
print("r2_test_ridge Score:", r2_test_ridge)
```

```python
lasso=Lasso(alpha=1)
lasso.fit(X_train, y_train)
print(lasso.intercept_)
print(lasso.coef_)
```
